```python
import requests
from bs4 import BeautifulSoup

# URL of the web page you want to scrape
url = 'https://excelchamps.com/blog/useful-macro-codes-for-vba-newcomers/#Download_the_PDF_File'

# Send a GET request to the URL
response = requests.get(url)

# Parse the content of the response with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all p tags and wp-block-preformatted tags
p_tags = soup.find_all('p')
preformatted_tags = soup.find_all(class_='wp-block-preformatted')

# Skip the first 5 p tags
useful_p_tags = p_tags[5:]

# Pair the useful p tags with the wp-block-preformatted tags
pairs = []
for p_tag, pre_tag in zip(useful_p_tags, preformatted_tags):
    description = p_tag.get_text().strip()
    code_snippet = pre_tag.get_text().strip()
    pairs.append((description, code_snippet))

# Annotate the data in a neat format
for i, (description, code_snippet) in enumerate(pairs, 1):
    print(f"Snippet {i}:\nDescription: {description}\nCode:\n{code_snippet}\n{'-'*40}")

# Optionally, you can save the annotated data to a file
with open('scraped_data.txt', 'w') as f:
    for i, (description, code_snippet) in enumerate(pairs, 1):
        f.write(f"Snippet {i}:\nDescription: {description}\nCode:\n{code_snippet}\n{'-'*40}\n")
```

```python
import json

# Function to parse the data
def parse_snippets(file_path):
    snippets = []
    with open(file_path, 'r') as file:
        content = file.read()

    # Split by the delimiter for each snippet
    raw_snippets = content.split('----------------------------------------')

    for raw_snippet in raw_snippets:
        if 'Description:' in raw_snippet and 'Code:' in raw_snippet:
            parts = raw_snippet.split('Code:')
            description = parts[0].replace('Description:', '').strip()
            code = parts[1].strip()

            snippets.append({
                "description": description,
                "code": code
            })

    return snippets

# Convert the data
snippets = parse_snippets('s')

# Save to JSON
with open('vba_snippets.json', 'w') as outfile:
    json.dump(snippets, outfile, indent=2)
```
